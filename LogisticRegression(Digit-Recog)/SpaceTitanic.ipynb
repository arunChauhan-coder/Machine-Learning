## Encoding

```python
from sklearn.preprocessing import LabelEncoder
col=[]
for i in df.columns:
    if df[i].dtype == 'O' or df[i].dtype == 'bool':
        if df[i].nunique() <3: # for bool
            label = LabelEncoder()
            df[i] = label.fit_transform(df[i])
        if df[i].nunique() > 2: # for multiple value
            col.append(i)
df = pd.get_dummies(df, columns= col)
            
            
```

```python
df.drop("Deck_A", axis =1 , inplace = True)
df.drop("HomePlanet_Earth",axis =1 , inplace = True)
df.drop("Destination_55 Cancri e",axis =1 , inplace = True)
#df.drop("Deck_G",axis =1 , inplace = True) due to multi colinearity
```

```python
x = df.drop("Transported", axis = 1)
y = df[["Transported"]]
```

## Outlier Treatment 

```python
for i in x:
    if x[i].dtype =='float64':
        sns.distplot(x[i])
        plt.show()
```

```python
for i in x:
    if x[i].dtype =='float64':
        sns.boxplot(x[i])
        plt.show()
```

```python
for i in x:
    if x[i].dtype =='float64':
        iqr = x[i].quantile(0.75)- x[i].quantile(0.25)
        LL = x[i].quantile(0.25) - 1.5*iqr
        UL = x[i].quantile(0.75)+1.5 *iqr
        
        # capping
        x[i] = np.where(x[i] > UL,UL, np.where(x[i] <LL,LL,x[i] ))
```

```python
for i in x:
    if x[i].dtype =='float64':
        sns.boxplot(x[i])
        plt.show()
```

## Splitting data into Train and Test set

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(sc_x, y, test_size=0.25, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Building Model

```python
from sklearn.linear_model import LogisticRegression
logit_ovr = LogisticRegression(multi_class='ovr')
logit_ovr.fit(x_train, y_train)
```

```python
# predict
y_pred_train = logit_ovr.predict(x_train)
y_pred_test = logit_ovr.predict(x_test)
```

## Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
print("Trainging Accuracy Score :", accuracy_score(y_train, y_pred_train))
print("****************"*10)
print("Test Accuracy Score :", accuracy_score(y_test, y_pred_test))
```

```python
print( classification_report(y_train, y_pred_train))
print("****************"*10)
print(classification_report(y_test, y_pred_test))
```

## Confusion Matrix

```python
print( confusion_matrix(y_train, y_pred_train))
print("****************"*10)
print(confusion_matrix(y_test, y_pred_test))
```

## ROC and AUC 

```python
from sklearn.metrics import roc_curve
fpr, tpr, thresholds = roc_curve(y_test, y_pred_test)
display(fpr[:10])
display(tpr[:10])
display(thresholds[:10])
```

```python
from sklearn.metrics import roc_auc_score
logit_roc_auc = roc_auc_score(y_test, y_pred_test)
logit_roc_auc
```

```python
plt.figure(figsize=(15,15))
plt.plot(fpr, tpr, label="ROC and AUC curve (area=%0.2f)" % logit_roc_auc)
plt.plot([0,1],[0,1], 'k--')
plt.xlim([0.0,1.0])
plt.ylim([0.0,1.0])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title("************Receiver Operating Characteristic with Area Under Curve************")
plt.legend(loc='lower right')
plt.show()
```

## Cross Validation

```python
# Cross Validation approach - K-Fold Method
from sklearn.model_selection import cross_val_score
training_accuracy = cross_val_score(logit_ovr, x_train, y_train, cv=10)
test_accuracy = cross_val_score(logit_ovr, x_test, y_test, cv=10)
print(training_accuracy)
print()
print(test_accuracy)
print()
print("Training Avg Accuracy", training_accuracy.mean())
print()
print("Test Avg Accuracy", test_accuracy.mean())
```
