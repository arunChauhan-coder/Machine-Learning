## Fixing Typos

```python
# there is ? in stalk-root

df.loc[df['stalk-root'] == '?', 'stalk-root'] = df['stalk-root'].value_counts().index[0]
```

## Remove Unnecessary col

```python
# viel-Type have only value in col that means it have zero variance  therefor drop it
```

```python
df.drop("veil-type",axis=1,inplace = True)
```

```python
df['habitat'].nunique()
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
lst = []
for col in df.columns:
    if df[col].nunique()<3:
        df[col] = label.fit_transform(df[col])
    else:
        lst.append(col)
        
df = pd.get_dummies(df, columns= lst)
```

```python
df.head()
```

```python
columns_to_drop = [
    'gill-attachment', 'cap-shape_b', 'cap-surface_f', 'cap-color_b', 'odor_a',
    'gill-color_b', 'stalk-root_b', 'stalk-surface-above-ring_f','stalk-surface-below-ring_f',
    'stalk-color-above-ring_b', 'stalk-surface-below-ring_f', 'stalk-color-below-ring_b',
    'veil-color_n', 'ring-number_n', 'ring-type_e', 'spore-print-color_b', 'population_a', 'habitat_d'
]
df.drop(columns_to_drop,axis = 1 , inplace = True)
```

```python
df.shape
```

```python
x= df.drop('class',axis=1)
```

```python
y = df[['class']]
```

```python
y.value_counts() # balanced
```

## Splitting Data into Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train,y_test = train_test_split(x, y, train_size=0.75, random_state=102)
```

## SVM

```python
from sklearn.svm import SVC

#kernel - linear
svm_linear = SVC(kernel='linear')
svm_linear.fit(x_train, y_train)
y_pred_train_linear = svm_linear.predict(x_train)
y_pred_test_linear = svm_linear.predict(x_test)

#kernel - sigmoid
svm_sigmoid = SVC(kernel='sigmoid')
svm_sigmoid.fit(x_train, y_train)
y_pred_train_sigmoid = svm_sigmoid.predict(x_train)
y_pred_test_sigmoid = svm_sigmoid.predict(x_test)

#kernel - poly
svm_poly = SVC(kernel='poly')
svm_poly.fit(x_train, y_train)
y_pred_train_poly = svm_poly.predict(x_train)
y_pred_test_poly = svm_poly.predict(x_test)

#kernel - rbf
svm_rbf = SVC(kernel='rbf')
svm_rbf.fit(x_train, y_train)
y_pred_train_rbf = svm_rbf.predict(x_train)
y_pred_test_rbf = svm_rbf.predict(x_test)
```

```python
# Evaluation 

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
print("Training Accuracy - Linear :", accuracy_score(y_train, y_pred_train_linear))
print("*************"*5)
print("Test Accuracy - Linear :", accuracy_score(y_test, y_pred_test_linear))
print("*************"*5)
print("Training Accuracy - sigmoid :", accuracy_score(y_train, y_pred_train_sigmoid))
print("*************"*5)
print("Test Accuracy - sigmoid :", accuracy_score(y_test, y_pred_test_sigmoid))
print("*************"*5)
print("Training Accuracy - poly :", accuracy_score(y_train, y_pred_train_poly))
print("*************"*5)
print("Test Accuracy - poly :", accuracy_score(y_test, y_pred_test_poly))
print("*************"*5)
print("Training Accuracy - rbf :", accuracy_score(y_train, y_pred_train_rbf))
print("*************"*5)
print("Test Accuracy - rbf :", accuracy_score(y_test, y_pred_test_rbf))
```

```python
from sklearn.model_selection import cross_val_score
train_accuracy1 = cross_val_score(svm_linear, x_train, y_train, cv=10, scoring='accuracy')
test_accuracy1 = cross_val_score(svm_linear, x_test, y_test, cv=10, scoring='accuracy')
train_accuracy1
```

## Voting

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logit = LogisticRegression()
logit.fit(x_train, y_train)
```

## Decesion Tree Classifier

```python
from sklearn.tree import DecisionTreeClassifier
dtree = DecisionTreeClassifier()
dtree.fit(x_train, y_train)
```

## KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier()
knn.fit(x_train, y_train)
```

## Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200, criterion='entropy', 
                            bootstrap=True, oob_score=False)
rf.fit(x_train, y_train)
```

## Hard Voting

```python
# Model building
estimators = [('SVM_Linear',svm_linear),('Logistic',logit),('Dtree', dtree),('KNN',knn),('RandomForestClassifier',rf)]
from sklearn.ensemble import VotingClassifier
voting_hard1 = VotingClassifier(estimators = estimators, voting='hard')
voting_hard1.fit(x_train, y_train)
```

```python
#Prediction
y_pred_train_voting = voting_hard1.predict(x_train)
y_pred_test_voting = voting_hard1.predict(x_test)
```

```python
#Accuracy
print("Training Accuracy - voting_hard :", accuracy_score(y_train, y_pred_train_voting))
print("*************"*5)
print("Test Accuracy - voting_hard :", accuracy_score(y_test, y_pred_test_voting))
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
v_train_accuracy = cross_val_score(voting_hard1, x_train, y_train, cv=10, scoring='accuracy')
v_test_accuracy = cross_val_score(voting_hard1, x_test, y_test, cv=10, scoring='accuracy')
print(np.round(np.mean(v_train_accuracy),2))
print()
print(np.round(np.mean(v_test_accuracy),2))
```

```python
# Cross Validation for each estimator speratly
```

```python
for estimator in estimators:
    train_scores = cross_val_score(estimator[1], x_train, y_train, scoring='accuracy', cv=10)
    test_scores = cross_val_score(estimator[1], x_train, y_train, scoring='accuracy', cv=10)
    print("Train Accuracy",estimator[0], np.round(np.mean(train_scores),2))
    print("******************"*5)
    print("Test Accuracy", estimator[0], np.round(np.mean(test_scores),2))
    print()
```

## Soft Voting

```python
# Model building
estimators = [('Logistic',logit),('Dtree', dtree),('KNN',knn),('RandomForestClassifier',rf)]
voting_soft = VotingClassifier(estimators = estimators, voting='soft')
voting_soft.fit(x_train, y_train)
```

```python
#Prediction
y_pred_train_voting = voting_soft.predict(x_train)
y_pred_test_voting = voting_soft.predict(x_test)
```

```python
#Accuracy
print("Training Accuracy - voting_soft :", accuracy_score(y_train, y_pred_train_voting))
print("*************"*5)
print("Test Accuracy - voting_soft :", accuracy_score(y_test, y_pred_test_voting))
```

```python
# Cross Validation
from sklearn.model_selection import cross_val_score
v_train_accuracy = cross_val_score(voting_soft, x_train, y_train, cv=10, scoring='accuracy')
v_test_accuracy = cross_val_score(voting_soft, x_test, y_test, cv=10, scoring='accuracy')
print(np.round(np.mean(v_train_accuracy),2))
print()
print(np.round(np.mean(v_test_accuracy),2))
```
