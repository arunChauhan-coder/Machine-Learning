## Dividing Cat and Num Column

```python
df["Partner"].nunique()
```

```python
cat_col =[]
num_col = []
for col in df.columns:
    if df[col].dtype == 'O' or df[col].nunique()<10:
        cat_col.append(col)
    else:
        num_col.append(col)
```

```python
cat_col 
```

```python
num_col 
```

```python
def check_classes(df):  
    dict = {}
    for i in list(df.columns):
        dict[i] = df[i].value_counts().shape[0]

    unq = pd.DataFrame(dict,index=["Unique Count"]).transpose().sort_values(by="Unique Count", ascending=False)
    return unq
```

```python
check_classes(df)
```

## Encoding

```python
for col in cat_col:
    if df[col].dtype!='O':
        
        df[col]=df[col].astype('O')
```

```python
from sklearn.preprocessing import LabelEncoder
col=[]
for i in cat_col:   
    if df[i].nunique() <3: # for bool
        label = LabelEncoder()
        df[i] = label.fit_transform(df[i])
    if df[i].nunique() > 2: # for multiple value
        col.append(i)
df = pd.get_dummies(df, columns= col)
```

```python
df.head()
```

```python
# To avoid multi colinearity drop one col

df.drop("MultipleConnections_No phone service", axis =1, inplace = True)
df.drop("InternetConnection_DSL", axis =1, inplace = True)
df.drop("OnlineSecurity_No", axis =1, inplace = True)
df.drop("DeviceProtectionService_No", axis =1, inplace = True)
df.drop("TechnicalHelp_No", axis =1, inplace = True)
df.drop("OnlineTV_No", axis =1, inplace = True)
df.drop("OnlineMovies_No", axis =1, inplace = True)
df.drop("Agreement_Month-to-month", axis =1, inplace = True)
df.drop("PaymentMethod_Bank transfer (automatic)", axis =1, inplace = True)
```

```python
x = df.drop("Churn",axis =1)
y = df[["Churn"]]
```

## Splitting data Train and Test

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

## Bagging Model

```python
from sklearn.ensemble import BaggingClassifier
bagging = BaggingClassifier()
bagging.fit(x_train, y_train)
```

```python
# predict
y_pred_train_bgg = bagging.predict(x_train)
y_pred_test_bgg = bagging.predict(x_test)
```

```python
# Evaluate 
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
```

```python
print(confusion_matrix(y_train, y_pred_train_bgg))
print()
print(confusion_matrix(y_test, y_pred_test_bgg))
```

```python
print(classification_report(y_train, y_pred_train_bgg))
print()
print(classification_report(y_test, y_pred_test_bgg))
```

```python
print(accuracy_score(y_train, y_pred_train_bgg))
print()
print(accuracy_score(y_test, y_pred_test_bgg))
```

```python
# Cross Validation approach
from sklearn.model_selection import cross_val_score
training_accuracy = cross_val_score(bagging, x_train, y_train, cv=10)
test_accuracy = cross_val_score(bagging, x_test, y_test, cv=10)
print(training_accuracy.mean())
print()
print(test_accuracy.mean())
```

## Grid Search CV

```python
from sklearn.model_selection import GridSearchCV
np.random.seed(42)
start = time.time()

param_dist = {'max_depth': [25,27,30],
              'min_samples_split':[3,4, 5, 6,8] ,
              'min_samples_leaf':[1, 2, 3,4,5],
              'bootstrap': [True, False],
              'max_features': ['auto', 'sqrt', 'log2', None],
              'criterion': ['gini', 'entropy']
              }

cv_rf = GridSearchCV(rf, cv = 10,
                     param_grid=param_dist, 
                     n_jobs = 7,
                     verbose = 3)

cv_rf.fit(x_train, y_train)
print('Best Parameters using grid search: \n', cv_rf.best_params_)
end = time.time()
print('Time taken in grid search: {0: .2f}'.format(end - start))
```
